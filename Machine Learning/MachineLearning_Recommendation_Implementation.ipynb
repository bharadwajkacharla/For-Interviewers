```python
dataset = pd.read_csv('radio_songs.csv')
dataset.head()
```

```python
# Dropping the column named "user"
radio_song_item = dataset.drop('user', 1)
radio_song_item.head()
```

```python
# Creating a placeholder dataframe listing song vs. song
radio_song = pd.DataFrame(index=radio_song_item.columns,columns=radio_song_item.columns)
radio_song.head()
```

```python
radio_song.fillna(0)
#radio_song = pairwise_distances(radio_song, metric='cosine')
radio_song
```

```python
#calculating the cosine_similarity
# Loop through the columns
for i in range(0,len(radio_song.columns)) :
    # Loop through the columns for each column
    for j in range(0,len(radio_song.columns)) :
      # Fill in placeholder with cosine similarities
      radio_song.iloc[i,j] = 1-cosine(radio_song_item.iloc[:,i],radio_song_item.iloc[:,j])
```

```python
radio_song
```

```python
# Create a placeholder items for closes neighbours to an item
radio_similar_songs = pd.DataFrame(index=radio_song.columns,columns=range(1,13))

# Loop through our similarity dataframe and fill in neighbouring item names
for i in range(0,len(radio_song.columns)):
    radio_similar_songs.iloc[i,:12] = radio_song.iloc[0:,i].sort_values(ascending=False)[:12].index
```

### 2. Conceptual questions:

##### a) Name 2 other similarity measures that you can use instead of cosine similarity above.


Similarity measures how much two objects are alike. Similarity measure in a data mining context is a distance with dimensions representing features of the objects. If distance is small, two objects are very similar where as if distance is large we will observe low degree of similarity.
There are lot of similarity distance measures.
- Cosine similarity
- Manhattan distance
- Euclidean distance
- Minkowski distance
- Jaccard similarity

##### Manhattan distance
    Manhattan distance is a metric in which the distance between two points is the sum of the absolute differences of their Cartesian coordinates. In a simple way of saying it is the total sum of the difference between the x-coordinates and y-coordinates.

#### b) What is needed to build a Content-Based Recommender system?




- Collaborative filtering gives recommendations based on other unknown users who have the same preference as a given user, but with content-based filtering, items are recommended on a feature-level basis
- The Content-Based Recommender system doesn't need any data about other users, since the recommendations are specific to the user. It simply helps you by identifying products that are similar to the product that a user likes. Considering that there's no need of data it can be scaled to large set of users. 
- The content based recommender system works by capture the specific interests of a user, and can recommend niche items that very few other users are interested in
- In Content-Based Recommender system, we match users based on the content they prefer. The attributes of the users and the products are very important
- Kernel density estimation (to help determine the density of data distribution) and cosine similarity (for measuring similarity between vectors) can be used to identify the recommendations for content-based recommender system

#### c) Name 2 methods to evaluate your recommender system.


There are two ways of evaluating the recommender system,
   - Accuracy and Error Based Methods
   - Ranking Based Methods

#### Accuracy and Error Based Methods:
   - Mean Absolute Error (MAE)
   - Mean Squared Error (MSE)
   - Root Mean Squared Error (RMSE)


##### Mean Squared Error (MSE)
    - The MSE is the average of the squared error. It is square of the average of the difference between the value predicted by the recommender and the actual value given by the user
    
   <img src='Screen Shot 2020-11-12 at 7.52.22 PM.png'>
   
##### Ranking Based Methods:
  - nDCG
  - Mean Reciprocal Rank
  - Average Precision
  - Spearman rank correlation evaluation

##### nDCG
     - nDCG has three parts. First is ‘CG’ which stands for Cumulative Gains. It deals with the fact that most relevant items are more useful than somewhat relevant items that are more useful than irrelevant items. It sums the items based on its relevancy, hence, the term cumulative.  
     - But CG doesn’t account for the position of the items on the list, changing the item's position won’t change the CG. To account for the item position we use the 'D' i.e., Discounted Gain.  
     - DG penalizes the items that appear lower in the list. A relevant item appearing at the end of the list is a result of a bad recommender system and hence that item should be discounted to indicate the bad performance of the model
     - DCG helps with the ranking but if we are comparing the different lists of the recommender. To normalize this ’n’ of nDCG, the third part, comes in to play.
